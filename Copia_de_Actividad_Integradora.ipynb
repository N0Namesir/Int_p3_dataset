```python
#Importar las librerías con las que trabajaremos
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importar el archivo csv con el que trabajaremos
url = "https://raw.githubusercontent.com/N0Namesir/Int_p3_dataset/main/Sales.csv"
df_bikes = pd.read_csv(url)
```

```python
#Realizar una copia del DataFrame en el que trabajaremos
bikes_sales= df_bikes.copy()
```

```python
#Guardar una copia de seguridad para comparaciones posteriores
df_backup= bikes_sales.copy()
```

```python
#Limpieza y procesamiento de datos
#1.Cargar los datos del DataFrame, df_bikes
print("\nDataFrame de ventas de bicicletas en una tienda Europea (primeras filas):\n", bikes_sales.head())#Muestra las primeras filas del Dataframe
```

```python
#Resume las columnas del DataFrame y sus tipos de datos
bikes_sales.info()
```

```python
#2.Eliminar columnas innecesarias para el análisis de datos
bikes_sales = bikes_sales.drop(columns=['Day', 'Month', 'Year'])  #Especifica el nombre de la columna a eliminar en el DataFrame, bikes_sales
```

```python
#Comprobar que las columnas innecesarias han sido eliminadas exitosamente
bikes_sales.info()
```

```python
#3.Manejar valores nulos o faltantes en el DataFrame
bikes_sales.isnull().sum()
```

```python
#Buscar registros duplicados en el DataFrame
duplicados= bikes_sales.duplicated()
print("Registros duplicados encontrados:", duplicados.sum())
```

```python
#Eliminar registros duplicados
df_clean = bikes_sales.drop_duplicates()
```

```python
#Confirmar si se eliminaron los registros duplicados
print("Duplicados en el original:", df_backup.duplicated().sum())  # Debe mostrar 1,000
print("Registro con duplicados eliminados:", df_clean.duplicated().sum())
```

```python
#5.Corregir Formatos
#Corregir tipos de datos de las columnas
#Corregir Date con tipo de dato "object" a "datetime"
df_clean['Date'] = pd.to_datetime(df_clean['Date'], format='%Y-%m-%d', errors='coerce')

#Corregir Age_Group, Customer_Gender, Country, State, Producto_Category, Sub_Category, Product, con tipo de dato "Object" a "category"
categorical_cols = ['Age_Group', 'Customer_Gender', 'Country', 'State',
                   'Product_Category', 'Sub_Category', 'Product']
df_clean[categorical_cols] = df_clean[categorical_cols].astype('category')

# Enteros pequeños a int16/int32 (si los valores lo permiten), para ocupar menos memoria
int_cols = ['Customer_Age', 'Order_Quantity']
df_clean[int_cols] = df_clean[int_cols].astype('int16')

# Convertir columnas monetarias a float32 (para decimales futuros)
money_cols = ['Unit_Cost', 'Unit_Price', 'Profit', 'Cost', 'Revenue']
df_clean[money_cols] = df_clean[money_cols].astype('float32')
```

```python
#Comprobar los cambios en los tipos de datos
df_clean.info()
```

```python
#Normalizar textos
text_cols = ['Customer_Gender', 'Country', 'State', 'Product_Category', 'Sub_Category', 'Product']#Selecciona las columnas a modificar
df_clean[text_cols] = df_clean[text_cols].apply(lambda x: x.str.strip().str.title())#Elimina espacione en blanco al inicio y al final, y convierte la primera letra de cada palabra a mayúscula
                                                                                          #Evita inconsistencias en los datos
```

```python
#Comprobar los cambios en el formato de los textos
df_clean.head()
```

```python
  # comienza la estadistica descriptiva
desc_stats = df_clean.describe().round(2)
print(f'Estadísticas descriptivas básicas:\n{desc_stats}')
```

```python
#media, mediana, desviacion pero solo con datos numericos
mean = df_clean.mean(numeric_only=True).round(2)
print(f'Media:\n{mean} \n')

medi = df_clean.median(numeric_only=True).round(2)
print(f'Mediana:\n{medi} \n')

dvc_stand = df_clean.std(numeric_only=True).round(2)
print(f'Desviación estándar:\n{dvc_stand} \n')
```

```python
grupo_categorico = df_clean.groupby('Product_Category').agg({
    'Revenue': ['sum', 'mean', 'median', 'std'],
    'Profit': ['sum', 'mean', 'median', 'std'],
    'Order_Quantity': ['sum', 'mean', 'median', 'std']
})

# fromatear para que tengan dos decimales
print('Estadísticas descriptivas por categoría de producto:')
print(grupo_categorico.to_string(formatters={col: '{:.2f}'.format for col in grupo_categorico.columns}))
print('\n')
```

```python
# creando una columna
# Crear nuevas columnas: margen de ganancia (%) y año de la venta
df_clean = df_clean.copy() # crear una copia explicita para convertir
df_clean['Profit_Margin_%'] = (df_clean['Profit'] / df_clean['Revenue']) * 100
df_clean['Profit_Margin_%'] = df_clean['Profit_Margin_%'].astype(float).round(2)


# Mostrar las primeras filas con las nuevas columnas
df_clean[['Date', 'Revenue', 'Profit', 'Profit_Margin_%']].head()
```
